## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data_no_UK = data[data['Country'] != 'United Kingdom']
data_no_UK
```

```python
data_no_UK['year'] = data_no_UK['InvoiceDate'].dt.year
data_no_UK['month'] = data_no_UK['InvoiceDate'].dt.month
data_no_UK['day'] = data_no_UK['InvoiceDate'].dt.day
data_no_UK
```

```python
data_no_UK = data_no_UK.loc[(data_no_UK['year'] == 2011) & (data_no_UK['month'] == 4)]
data_no_UK
```

```python
data_no_UK_grouped = data_no_UK.groupby('Country').sum()
data_no_UK_grouped.head()
data_no_UK_grouped = data_no_UK_grouped.reset_index()
data_no_UK_grouped
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=data_no_UK_grouped.Country,
    y=data_no_UK_grouped.Quantity,
    name='Quantity',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=data_no_UK_grouped.Country,
    y=data_no_UK_grouped.Revenue,
    name='Revenue',
    marker_color='lightsalmon'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
data_france = data[data['Country'] == 'France']
data_france['year'] = data_france['InvoiceDate'].dt.year
data_france['month'] = data_france['InvoiceDate'].dt.month
data_france['day'] = data_france['InvoiceDate'].dt.day

data_france
```

```python
data_france = data_france.loc[(data_france['year'] == 2011) & (data_france['month'] >= 1) & (data_france['month'] <= 5)]
data_france
```

```python
data_france = data_france.groupby('month').sum()
data_france = data_france.reset_index()
data_france.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Line(
    x=data_france.month,
    y=data_france.Quantity,
    name='Quantity',
    marker_color='indianred'
))
fig.add_trace(go.Line(
    x=data_france.month,
    y=data_france.Revenue,
    name='Revenue',
    marker_color='lightsalmon'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()


```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
data_party = data[data['Description'] == 'PARTY BUNTING']
data_party
```

```python
data_party = data_party.groupby('Country').mean()
data_party = data_party.reset_index()
data_party.head()
```

```python
fig = px.scatter(data_party, x="Quantity", y="UnitPrice", color="Country")
fig.show()
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

## 7. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
agg_func = {'InvoiceNo':'nunique',
            'Quantity':'sum',
            'UnitPrice':'mean',
            'Revenue':'sum',
            'CustomerID':'nunique'}

products = uk.groupby('Description').agg(agg_func)
products = products.reset_index()
products
```

```python
fig = px.scatter(products, x="InvoiceNo", y="CustomerID")
fig.show()
```
